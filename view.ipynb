```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, save_img
from pathlib import Path
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("index.csv")
```

```python
def tile(df, m, n, root=".", filename_col="filename", label_col="label", figsize=(24, 24)):
    fig, axes = plt.subplots(m, n, figsize=figsize)
    for i in range(m*n):
        ax = axes[i // n][i % n]
        j = np.random.randint(len(df))
        ax.axis("off")
        path = df[filename_col][j]
        label = df[label_col][j]
        ax.set_title(f"{j}. {path.split('/')[-1]} {label}", fontsize=18)
        im = load_img(path)
        ax.imshow(im)
```

```python
tile(df, 4, 4)
```
